# Assignment 1

In this assignment, you'll be working with messy medical data and using regex to extract relevant infromation from the data. 

Each line of the `dates.txt` file corresponds to a medical note. Each note has a date that needs to be extracted, but each date is encoded in one of many formats.

The goal of this assignment is to correctly identify all of the different date variants encoded in this dataset and to properly normalize and sort the dates. 

Here is a list of some of the variants you might encounter in this dataset:
* 04/20/2009; 04/20/09; 4/20/09; 4/3/09
* Mar-20-2009; Mar 20, 2009; March 20, 2009;  Mar. 20, 2009; Mar 20 2009;
* 20 Mar 2009; 20 March 2009; 20 Mar. 2009; 20 March, 2009
* Mar 20th, 2009; Mar 21st, 2009; Mar 22nd, 2009
* Feb 2009; Sep 2009; Oct 2010
* 6/2008; 12/2009
* 2009; 2010

Once you have extracted these date patterns from the text, the next step is to sort them in ascending chronological order accoring to the following rules:
* Assume all dates in xx/xx/xx format are mm/dd/yy
* Assume all dates where year is encoded in only two digits are years from the 1900's (e.g. 1/5/89 is January 5th, 1989)
* If the day is missing (e.g. 9/2009), assume it is the first day of the month (e.g. September 1, 2009).
* If the month is missing (e.g. 2010), assume it is the first of January of that year (e.g. January 1, 2010).
* Watch out for potential typos as this is a raw, real-life derived dataset.

With these rules in mind, find the correct date in each note and return a pandas Series in chronological order of the original Series' indices.

For example if the original series was this:

    0    1999
    1    2010
    2    1978
    3    2015
    4    1985

Your function should return this:

    0    2
    1    4
    2    0
    3    1
    4    3

Your score will be calculated using [Kendall's tau](https://en.wikipedia.org/wiki/Kendall_rank_correlation_coefficient), a correlation measure for ordinal data.

*This function should return a Series of length 500 and dtype int.*

```python
def date_sorter():


    import pandas as pd
    from datetime import datetime
    import datetime as dt
    import numpy as np
    import re
    import math

    doc = []
    with open('dates.txt') as file:
        for line in file:
            doc.append(line)

    df = pd.Series(doc)


    mylist = []

    for f in df:

        # fish out some difficult numbers causing trouble later
        f = f.replace('9079', "")
        f = f.replace('9080', "")
        f = f.replace('7787', "") 
        f = f.replace('36/308', "")  
        f = f.replace('Janaury', "January") 
        f = f.replace('Decemeber', "December")
        f = f.replace('sSince', "") 
        f = f.replace('In 1980', "1980")  


        c = re.findall(r"\d{1,2}[/-]\d{1,2}[/-]\d{1,4}", f)
        d = re.findall(r"[A-Z][a-z]+\s+\d{1,2}[,/-]\s+\d{4}|[A-Z][a-z]+\s+\d{2}\s+\d{4}", f)
        e = re.findall(r"\d{1,2}\s+[A-Z][a-z]+\s+\d{4}", f)
        g = re.findall(r"(\d{4}|\d{1,2}[/]\d{4}|[A-Z][a-z]+\s+\d{4})", f)
        #print(d)06/1973
        mylist.append([c,d,e,g])
        # July 26, 1978
    df1 = pd.DataFrame(mylist)


    def guess_date(string):
        for fmt in ["%Y/%m/%d", "%d-%m-%Y", "%m-%d-%Y", "%Y%m%d", "%m/%d/%Y","%d/%m/%Y",
                "%y/%m/%d", "%m/%y" , "%m/%Y","%m-%d-%y", "%d-%m-%y", "%y%m%d", "%m/%d/%y","%d/%m/%y",
               '%Y','%b %d, %Y','%b %d, %Y','%B %d, %Y','%B %d %Y','%m/%d/%Y','%m/%d/%y',
                '%b %Y','%B%Y','%b %d,%Y']:
            try:
                return datetime.strptime(string, fmt).date()
                #return datetime.datetime.strptime(string, fmt).date()
            except ValueError as err:
                    pass 

    def guess_date1(string):
        for fmt in ["%Y"]:
            try:
                return datetime.strptime(string, fmt).date()
                #return datetime.datetime.strptime(string, fmt).date()
            except ValueError as err:
                    pass 

        #print(ValueError(string))

    def guess_date2(string):
        for fmt in ["%d %b %Y", "%d %B %Y", "%d/%m/%y", "%B %Y", "%b %Y", "%m/%Y", "%Y"]:
            try:
                return datetime.strptime(string, fmt).date()
                #return datetime.datetime.strptime(string, fmt).date()
            except ValueError as err:
                    pass 

    #print(ValueError(string))

    one = []
    for i in df1[0]:
        t = ("".join(i))
        d = (guess_date(t))
        row_str = str(d)
        if row_str == "None":
            row_str = 0
            one.append(row_str)

        else:
            d1 = int((d.strftime('%Y%m%d')))
            one.append(d1)

    two = []
    for i in df1[1]:
        t = ("".join(i))
        d = (guess_date(t))
        row_str = str(d)
        if row_str == "None":
            row_str = 0
            two.append(row_str)

        else:
            d1 = int((d.strftime('%Y%m%d')))
            two.append(d1)


    three = []
    for i in df1[2]:
        t = ("".join(i))
        d = (guess_date2(t))
        row_str = str(d)
        if row_str == "None":
            row_str = 0
            three.append(row_str)

        else:
            d1 = int((d.strftime('%Y%m%d')))
            three.append(d1)


    four = []
    for i in df1[3]:
        t = ("".join(i))
        d = (guess_date2(t))
        row_str = str(d)
        if row_str == "None":
            row_str = 0
            four.append(row_str)

        else:
            d1 = int((d.strftime('%Y%m%d')))
            four.append(d1)


    df_n = pd.DataFrame(list(zip(one, two,three, four)), columns =['c', 'd',"e","g"])
    df_n.replace(to_replace=0,value=np.nan, inplace=True)

    df_n[4] = df_n["c"].combine_first(df_n["e"])
    df_n[4] = df_n[4].combine_first(df_n["d"])
    df_n[4] = df_n[4].combine_first(df_n["g"])

    df_n.sort_values(by = 4, inplace=True)
    df_n.reset_index(inplace=True)

    return df_n["index"]
```
